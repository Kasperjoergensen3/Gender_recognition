```python
model_age = tf.keras.models.load_model('./Models/Age')
with open('./Histories/Age/fold_1', "rb") as file_pi:
    history_age = pickle.load(file_pi)
fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 5))
#  "Accuracy"
axes[0].plot(history_age['accuracy'])
axes[0].plot(history_age['val_accuracy'])
axes[0].set(xlabel='Epochs', ylabel='Accuracy')
axes[0].set_title('Model Accuracy for Age Model')
axes[0].legend(['train', 'validation'], loc='upper left')
# "Loss"
axes[1].plot(history_age['loss'])
axes[1].plot(history_age['val_loss'])
axes[1].set(xlabel='Epochs', ylabel='Loss')
axes[1].set_title('Model Loss for Age Model')
axes[1].legend(['train', 'validation'], loc='upper left')
plt.savefig('./Results/age_learning_curve.png')
```

```python
model_multitask = tf.keras.models.load_model('./Models/Multitask')
with open('./Histories/Multitask/fold_1', "rb") as file_pi:
    history_multitask = pickle.load(file_pi)
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(10, 10))
#  "Accuracy"
axes[0,0].plot(history_multitask['gender_accuracy'])
axes[0,0].plot(history_multitask['val_gender_accuracy'])
axes[0,0].set(xlabel='Epochs', ylabel='Accuracy')
axes[0,0].set_title('Gender Classification Accuracy for Multitask Model')
axes[0,0].legend(['train', 'validation'], loc='upper left')
# "Loss"
axes[0,1].plot(history_multitask['gender_loss'])
axes[0,1].plot(history_multitask['val_gender_loss'])
axes[0,1].set(xlabel='Epochs', ylabel='Loss')
axes[0,1].set_title('Gender Classification Loss for Multitask Model')
axes[0,1].legend(['train', 'validation'], loc='upper left')
#  "Accuracy"
axes[1,0].plot(history_multitask['age_accuracy'])
axes[1,0].plot(history_multitask['val_age_accuracy'])
axes[1,0].set(xlabel='Epochs', ylabel='Accuracy')
axes[1,0].set_title('Age Classification Accuracy for Multitask Model')
axes[1,0].legend(['train', 'validation'], loc='upper left')
# "Loss"
axes[1,1].plot(history_multitask['age_loss'])
axes[1,1].plot(history_multitask['val_age_loss'])
axes[1,1].set(xlabel='Epochs', ylabel='Loss')
axes[1,1].set_title('Age Classification Loss for Multitask Model')
axes[1,1].legend(['train', 'validation'], loc='upper left')
fig.tight_layout()
plt.savefig('./Results/multitask_learning_curve.png')
```
